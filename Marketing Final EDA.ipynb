### Import the necessary packages

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Pull in the demographics excel file

```python
df_demo = pd.read_excel("/Users/jasoneisele/Desktop/Mktg. Analytics_Demographics_Quick Report.xlsx",header = 0)
```

```python
df_demo.head()
```

```python
df_demo.columns
```

```python
print("The current shape of our dataframe is {}".format(df_demo.shape[0]) + " rows and {}".format(df_demo.shape[1]) + " columns")
```

```python
#Flip the rows and columns
df_demo = df_demo.transpose()
```

```python
print("The new shape of our dataframe is {}".format(df_demo.shape[0]) + " rows and {}".format(df_demo.shape[1]) + " columns")
```

```python
df_demo.index.astype('str')
```

```python
df_demo.reset_index(inplace = True)
```

```python
df_demo.head()
```

```python
df_demo = df_demo.rename(index = str, columns = {'index':"Area"})
```

```python
# Show all columns in dataframe
pd.set_option('display.max_columns', None)
```

```python
df_demo.head()
```

```python
texas = df_demo.iloc[[1,3,4,5],:]
```

```python
texas
```

```python
missouri = df_demo.iloc[[2,6,7],:]
```

```python
missouri
```

```python
state_level = df_demo.iloc[[0,1,2],:]
```

```python
state_level
```

```python
sns.barplot(x = 'Area', y = 'Population (Pop), 2018',data = state_level, palette="Blues_d")
plt.title("Population 2018")
plt.xlabel("US Compared to TX and Missouri")
plt.ylabel("Billions")
```

```python
sns.barplot(x = 'Area', y = 'Urban Population, 2018', data = texas, palette="Blues_d")
plt.title("Urban Population 2018")
plt.xlabel("Area")
plt.xticks(rotation = 45)
```
